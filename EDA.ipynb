```python
import plotly.express as px
```

```python
data = pd.read_csv("adult.csv")
data.head()
```

```python
grp_data = data.groupby('workclass')['capital-gain'].mean()
px.pie(grp_data, grp_data.index, grp_data.values, hole = 0.8)
```

## People from self emp inc and not inc are gaining more compared to other workclass people. 

```python
grp_data = data.groupby('occupation')['capital-gain'].mean()
px.pie(grp_data, grp_data.index, grp_data.values, hole = 0.2)
```

## On an average Prof-speciality and Exec-managerial gains more salary.

```python
df = data.loc[data['capital-gain'] < 50000]
px.box(df, x='workclass', y='capital-gain', color='salary')
```

### There are few outliers in the dataset and so we have removed the points.

```python
grp_data = data.groupby('country')['capital-gain'].mean().nlargest(10)
px.bar(grp_data, x = grp_data.index, y=grp_data.values, labels={'x':'country', 'y':'capital-gain'})
```

```python
#px.pie(data,names='race', values='capital-gain')
grp_data = data.groupby('race')['capital-gain'].mean().nlargest(5)
```

```python
px.pie(grp_data, grp_data.index, grp_data.values, hole = 0.05)
```
